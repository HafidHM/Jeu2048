Fonction renvoie une grid avec des 0 de taille size*size

```python
def init_grid(size):
    l = []
    ll=[]
    for i in range(size):
        l.append(0)
    for i in range(size):
        ll.append(l)
    return ll
```

Inialisation d'une partie

```python
 def init_game(size):
    game = GameEngine(size)
    return game
```

Fonction retourne la case max d'une grid 

```python
def maximum(board):
    return np.max(board)
```

Fonction teste si la grid contient une case de valeur 16 ou plus

```python
def is_16(board):
    if 16 in board or maximum(board)>16:
        return True
    else:
        return False
```

Fonction teste si la grid contient une case de valeur 32 ou plus

```python
def is_32(board):
    if 32 in board  or maximum(board)>32:
        return True
    else:
        return False
```

Fonction teste si on a gagné 

```python
def is_Win(grid):
    if is_32(grid):
        return True
    else:
        return False
```

Fonction permet de prendre  la décision d'arreter le jeu 

```python
def continue_playing(grid):
    if (np.any(grid == 0)):       
        return True
    else:
        for i in range(3):
            if (grid[i][0]==grid[i][1]) or (grid[i][1]==grid[i][2]):
                return True
            if (grid[0][i]==grid[1][i]) or (grid[1][i]==grid[2][i]):
                return True
        else:
            return False
```

Fonction permet d'affiche une grid 

```python
def Affiche_Grid(grid,size):
    str_temp = ""
    print(" ============== ")  
    for i in range(size):
        print("|              |")
        print("| ",end='')
        for j in range(size):
            if grid[i][j] == 0:
                print("%3s"%str_temp,end="")
            else:
                print("%3d"%grid[i][j],end ="")
        print("    |")
    print("|              |")
    print(" ============== ")
```

Fonction permet d'ajouter 2 ou 4 sur la grid d'une maniere aleatoirement et avec une probabilité très élevée d'avoire 2 
que d'avoir 4 

```python
def next_grid(grid,game):
    liste=position=[]
    liste = game.list_of_next_random(grid)
    for i in range(len(liste[0])):
        position.append(i)
    pos = np.random.choice(np.array(np.random.choice(position, len(position),p = liste[1])))
    return liste[0][pos] 
```

Fonction permet de proposer 4 directions au plus (Haut, Bas, Gauche, Droite)

```python
def choice_move(grid,game): 
    liste= []
    l=[]
    for i in range(len(grid)):
        for j in range(len(grid)):
            l.append(grid[i][j])
        liste.append(l)
        l = []
    return game.list_of_next_choices(liste) 
```

Retourne le maximum d'un plateau pour chaque direction

```python
def simuler_les_maximums(boards, nombre_d_experiences):
    maxs_partie = []
    experiences=[]
    for i in range(0, nombre_d_experiences):
        for plateau in boards:
            maxs_partie.append(maximum(plateau))
        experiences.append(maxs_partie)
        maxs_partie=[]
    return experiences
```

Choisir une direction (naive Strategie)

```python
def choisir_direction(maximums,choix):
    return choix[np.argmax(maximums)]
```

Retourne le Score d'une partie

```python
def score(board):
    return np.sum(board)
```

# Programme principale

```python
if __name__ == '__main__':
    #size de la Grid 3x3
    size = 3
    
    #Inialisation de la grid et le jeu
    game = init_game(size)
    grid = init_grid(size)
    
    #Inserer aleatoirement deux valeurs dans la grid 
    grid = next_grid(grid,game)
    grid = next_grid(grid,game)
    #Afficher la grid
    Affiche_Grid(grid,size)

    while continue_playing(grid) == True: 
        #Lster les choix possibles
        choix = choice_move(grid,game)
        #Rtourner les max de chaque possibilite
        maximums= simuler_les_maximums(choix, 1)
        #Prendre une direction parmi les choix possibles 
        grid = choisir_direction(maximums,choix)
        print("Score: "+str(score(grid)))
        Affiche_Grid(grid,size)
        #Inserer 2 ou 4 dans la grid
        grid = next_grid(grid,game)
        Affiche_Grid(grid,size)
    
    #Test l'etat de la grid
    if is_Win(grid):
        print("Win !!")
        print("Score: "+str(score(grid)))
    else:
        print("Game Over")
```
